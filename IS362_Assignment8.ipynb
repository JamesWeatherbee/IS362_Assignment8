```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

Load auto-mpg.data into a pandas DataFrame

```python
df = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header=None)
pd.set_option('display.max_columns', None)
```

Assign column names to the df

```python
column_name = ['mpg',
          'cylinders',
          'displacement',
          'horsepower',
          'weight',
          'acceleration',
          'model_yr',
          'origin',
          'car_name']
df.columns = column_name
df
```

Replace occurances of '?' with NaN and covert the horsepower column to_numeric

```python
df.horsepower = df.horsepower.replace('?', np.nan)
df.horsepower = pd.to_numeric(df.horsepower)
df
```

Rename the origin column's data to country names

```python
origins = {1: 'USA', 
          2: 'Asia',
          3: 'Europe'}

df.origin = df.origin.map(origins)
df
```

Create a bar graph that shows the distribution of cylinders

```python
a = sns.countplot(x="cylinders", hue='origin' ,data=df)
plt.xlabel("Cylinders",size = 20)
plt.ylabel("Count",size = 20)
```

Create a scatterplot that shows the relationship between horsepower and weight

```python
b = sns.regplot(x="horsepower", y="weight", data=df, color='y')
plt.title('Horsepower Vs. Weight', fontsize=20)
plt.xlabel("Horsepower",size = 15)
plt.ylabel("Weight",size = 15)
```

What is the relationship between horsepower and car origin over the years?

```python
c = sns.lmplot(x='model_yr', y='horsepower', hue='origin', data=df)
plt.title('Horsepower over Time for each Country', fontsize=22)
plt.xlabel('Model Year', fontsize=15)
plt.ylabel('Horsepower', fontsize=15)
```
